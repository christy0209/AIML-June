### Combining 2 dataframes

```python
df1 = pd.DataFrame([[1,25,'F'],[2,26,'M'],[3,24,'F'],[4,27,'F'],[7,23,'F'],[8,20,'F'],[9,29,'M'],[10,21,'M']],columns=['ID','Age','Gender'])
```

```python
df1
```

```python
df2 = pd.DataFrame([[1,'A'],[2,'B'],[3,'C'],[4,'D'],[5,'E'],[6,'F'],[7,'G'],[8,'H']],columns=['ID','Name'])
```

```python
df2
```

### Merge is the function used for SQL joins in Pandas

```python
inner_join=pd.merge(df1,df2,how="inner",on="ID")
inner_join
```

```python
left_join=pd.merge(df1,df2,how="left",on="ID")
left_join
```

```python
right_join=pd.merge(df1,df2,how="right",on="ID")
right_join
```

```python
outer_join=pd.merge(df1,df2,how="outer",on="ID")
outer_join
```

## Concat function used to merge dataframes by index in row level and column level

### concat is accepting list of dataframes as parameter

```python
#row level
pd.concat([df1,df2],axis=0)
```

```python
#column level
pd.concat([df1,df2],axis=1)
```

```python
inner_join
```

```python
df3 = pd.DataFrame([[1,'A'],[2,'B'],[3,'C'],[4,'D'],[5,'E'],[6,'F'],[7,'G'],[8,'H'],[9,'I']],columns=['Name_ID','Name'])
```

```python
#column level
pd.concat([df1,df3],axis=1)  #we are passing the list of dataframes to be combined, also we need to specify the axis of combination
```

```python
df3
```

```python
df1
```

### Using Merge function with 2 key columns

```python
pd.merge(df1,df3,how="inner",left_on = 'ID',right_on='Name_ID')
```

```python
pd.merge(df1,df3,how="left",left_on = 'ID',right_on='Name_ID')
```

```python
pd.merge(df1,df3,how="right",left_on = 'ID',right_on='Name_ID')
```

```python
pd.merge(df3,df1,how="outer",left_on = 'Name_ID',right_on='ID')
```
